```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def calories_out(calories_in, mean_calories_in, m):
    return calories_in - mean_calories_in * m
```

```python
calories_in_data = []
calories_out_data = []
m = 0.9
d = 30 # not needed?
t = 1000

def update(m,d):

    calories_in_today = 1200 + np.random.normal(0, 1500) # get calories in from user input

    calories_in_data.append(calories_in_today) # get calories in from user input

    # compute mean calories in untl 60 past iterations
    mean_calories_in = np.mean(calories_in_data[-d:])

    # compute calories out
    calories_out_today = calories_out(calories_in_data[-1], mean_calories_in, m)
    calories_out_data.append(calories_out_today)
```

```python
for i in range(t):
    update(m,d)
```

```python
# plot data
plt.plot(calories_out_data)
plt.show()
```

```python
# plot sum of calories out
plt.plot(np.cumsum(calories_out_data))
```
