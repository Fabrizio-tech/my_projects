```python
import numpy as np
import matplotlib.pyplot as plt

def calories_out(cal_in, avg_cal_in, change_in_calories, k, s):
    metabolism = avg_cal_in * k
    # cap on metabolism, can't go below 1800 and above 5000
    if metabolism < 1800:
        metabolism = 1100
    if metabolism > 5000:
        metabolism = 5000

    movement = s

    random_diet = np.random.normal(0, change_in_calories)

    print('metabolism:', metabolism)
    print('movement:', movement)
    print('total calories out:', cal_in + random_diet - metabolism - movement)
    return cal_in + random_diet - metabolism - movement, metabolism, cal_in + random_diet

cal_in = 2000
avg_cal_in = 2000
change_in_calories = 200
k = 0.9
s = 200

T = 600

calories = []
metabolism = []
calories.append(avg_cal_in)
metabolism.append(avg_cal_in * k)
avg_cal_in_data = []
avg_cal_in_data.append(avg_cal_in)
actual_cals_in = []
for t in range(T):

    # compute the new avg_cal_in the last 50 days 
    avg_cal_in = np.mean(actual_cals_in[-20:])

    current_calories, meta, actual_cal_in = calories_out(cal_in, avg_cal_in, change_in_calories, k, s)
    calories.append(current_calories)
    actual_cals_in.append(actual_cal_in)
    metabolism.append(meta)
    avg_cal_in_data.append(avg_cal_in)

plt.plot(calories)
plt.plot(metabolism)
plt.plot(avg_cal_in_data)
plt.legend(['calories out', 'metabolism', 'avg_cal_in'])

plt.xlabel('Time')
plt.ylabel('Calories gained or lost')
```
